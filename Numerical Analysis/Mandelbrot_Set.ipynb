```julia
using PyPlot
```

```julia
function mandelbrot_iteration(c, max_iter=200)
    z = c
    for k = 1:max_iter
        if abs(z) > 2.0
            return log(k)
        end
        z = z^2 + c
    end
    return 0
end
```

```julia
compute_mandel(lower_left, upper_right, resolution=500, max_iter=200) = 
    [ mandelbrot_iteration(complex(j,i), max_iter)
        for i = linspace(upper_right.im, lower_left.im, resolution), 
            j = linspace(lower_left.re, upper_right.re, resolution)]
```

```julia
(lower_left, upper_right) = (-2-1im, 1+1im)

@time mandelbrot_set = compute_mandel(lower_left, upper_right, 1600, 3000);
```

```julia
using PyPlot

figure(figsize=(20,20))
axis("off")


imshow(mandelbrot_set, cmap="hot");
```

```julia
(lower_left, upper_right) = (-0.7462+0.1103im, -0.7460+0.1105im)
@time mandelbrot_set = compute_mandel(lower_left, upper_right, 1600, 3000)
figure(figsize=(20,20))
axis("off")
imshow(mandelbrot_set, cmap="hot");
```
